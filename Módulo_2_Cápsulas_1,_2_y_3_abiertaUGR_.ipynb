### **2.1. Datos de expresin gentica**

Los denominados datos de expresin gentica son muy populares en los anlisis de transcriptmica. Hoy da, este tipo de datos se obtienen con las denominadas tecnologas de secuenciacin de ARN (RNA-Sequencing o RNA-Seq). Estas tecnologas permiten identificar secuencias de ARN en una muestra celular y cuantificar su abundancia. Es decir, identificar qu genes se expresan en la muestra en ese instante y cul es su grado de expresin. Adems de cuantificar la expresin de genes, el anlisis de estos datos permite identificar nuevas secuencias transcritas a partir de ADN, identificar mecanismos de *splicing* alternativo o detectar expresin especfica de alelo, entre otros.  Adems, estas tecnologas permiten caracterizar no slo RNA mensajero (mRNA), sino tambin otros tipos de RNAs como los RNAs que no codifican protenas (los llamados RNAs no codificantes o *non-coding RNAs*, ncRNAs) que incluyen los lncRNAs y los miRNAs, entre otros.  

El resultado de estos experimentos suele ser una **matriz de expresin** *X* de tamao *N  M*, siendo *N* los genes y *M* las muestras, y cada valor de la matriz ${x_{i,j}}$ representa el valor de expresin del gene *i* en la muestra *j*.

Una caracterstica de estos datos es que *N << M*  genes suelen ser del orden de decenas de miles y los casos normalmente de unas pocas decenas. 

Puedes consultar ms detalles sobre la secuenciacin de ARN en los siguientes enlaces: 
- http://cshprotocols.cshlp.org/content/early/2015/04/11/pdb.top084970.abstract
- https://www.nature.com/articles/nrg2484
- https://training.galaxyproject.org/training-material/topics/transcriptomics/

Hay muchos pasos involucrados en el anlisis de datos de RNA-Seq antes de obtener la matriz de expresin. Tpicamente, este proceso comienza con el procesamiento de lecturas (*reads*), que se alinean contra un genoma de referencia para cuantificar el nmero de secuencias de ARN asociadas a cada posicin del genoma. Como se conoce la posicin en el genoma de un amplio repertorio de genes, se suele indicar que estas tcnicas permiten cuantificar el grado de expresin de cada gen en una muestra. Esta informacin se dispone en una matriz numrica sobre la que podemos realizar anlisis estadsticos y computacionales. En nuestro caso, y como aproximacin inicial al problema, partiremos directamente de las matrices que cuantifican el nmero de lecturas asociadas a cada gen y nos centraremos en el anlisis de estas matrices.

## **1. PREPROCESAMIENTO Y NORMALIZACIN DE LOS DATOS DE TCGA** <a id='preprocessing'></a>





El **preprocesamiento** de datos es una etapa de todo proceso computacional  donde los datos se preparan para su posterior tratamiento y anlisis. Esta etapa incluye cualquier tipo de transformacin, re-estructuracin, filtrado, eliminacin de ruido,  manejo de valores faltantes, etc. En machine learning, este tipo de transformaciones sobre los datos suele denominarse manipulacin de datos o **Data *wrangling***. Algunas formas de preprocesamiento habituales en datos micos son el filtrado de datos de mala calidad, filtrado de genes con poca variacibilidad, imputacin o filtrado de casos/genes con valores faltantes, etc.  

La **normalizacin** es un proceso de transformacin de los datos con el que se pretende estadarizar o eliminar efectos debidos a variaciones en medidas o experimentales para el posterior de estos datos. Por ejemplo, es habitual transformar el valor de variables numricas continuas de distinta magnitud a la escala $[0,1]$ para que puedan combinarse o compararse entre s. 
Se suele tambin aplicar una tranformacin a escala logartmica, ya que las propiedades de las distribuciones logartmicas resultan ms convenientes para representar esta informacin y facilita anlisis posteriores. 

Las tcnicas de normalizacin permiten corregir en parte la variabilidad o ruido inherente a las tcnicas experimentales en ciencias -micas. De este modo, estas tcnicas resultan fundamentales antes de combinar datos obtenidos en distintos experimentos, incluso si han sido producidos en el mismo laboratorio, por el mismo equipo tcnico y utilizando los mismos instrumentos. 

La visualizacin de los datos obtenidos resultar fundamental para el denominado **Control de Calidad**, etapa en la que se identifican patrones anormales que sealan que existen sesgos experimentales (*batch effects*) que no han sido corregidos  por los mtodos de normalizacin.  

En este *notebook* proponemos una serie de pasos para el preprocesamiento y normalizacin de los datos de expresin gentica de TCGA en lnea con lo propuesto en el [artculo de investigacin original](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4580370/) a travs del que dichos datos se hicieron pblicos en la comunidad cientfica. Estos pasos son los siguientes: 
1. Normalizacin del nmero de copias de RNAs con el mtodo [RSEM](https://bmcbioinformatics.biomedcentral.com/articles/10.1186/1471-2105-12-323) (ya aplicado sobre los datos descargados) y transformacin logartmica. 
2. Centrado del valor de expresin de los genes en su mediana.
3. Seleccin de los 1500 genes con mayor variabilidad. 
4. Cambios en el nombre de las muestras para que coincidan con los identificadores de muestra de las tablas con informacin clnica.

###2.2. Reduccin de dimensionalidad

Recordar que en este tipo de estudios se suele tener decenas de miles de variables cuantificadas en unas pocas decenas de muestras. Por tanto, es frecuente tambin el uso de tcnicas de reduccin de dimensionalidad como el anlisis de componentes principales (PCA) mediante la cual se sustituyen las observaciones originales por *n* combinaciones lineales, siendo *n* mucho menor las dimensiones originales. Se seleciona *n* componentes principales de forma que representen una proporcin razonable de la variacin total.

La funcin de R `prcomp` calcula las componentes principales de una matriz dada como argumento. El siguiente cdigo ilustra como identificar sobre la submatriz `milquinientosgenesdata[1:50,1:20]` los perfiles genticos de mxima varianza. Primero, consideramos que cada gen es una observacin y las muestras son las variables.

````
%%R
# 1- PCA aplicado sobre los genes
library(factoextra)
pca <- prcomp(milquinientosgenesdata[1:50,1:20])
fviz_eig(pca)

````

Este grfico representa cunta variabilidad est explicada por cada una de las componentes principales calculadas. En este caso, comprobamos que el 45% de la varianza en los datos est explicada por las dos primeras componentes principales. El siguiente grfico muestra cmo se representan los genes en un espacio definido por estas dos componentes principales


````
%%R
library(scales)
fviz_pca_ind(pca,
             col.ind = "cos2", # Color by the quality of representation
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
)
````

Ahora, procedemos de igual modo pero consideramos cada muestra una observacin, y cada gen una variable. Para ello, trasponemos la matriz anterior con la funcin `t`. El resto del cdigo no cambia

````
%%R
# 2- PCA aplicado sobre las muestras (transponiendo la matriz original con la funcion t)
pca <- prcomp(t(milquinientosgenesdata[1:50,1:20]))
fviz_eig(pca)
````

````
%%R
fviz_pca_ind(pca,
             col.ind = "cos2", # Color by the quality of representation
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
)
````

Estas representaciones grficas nos permiten realizar un primer acercamiento a los datos y a su distribucin en el espacio, identificando de forma preliminar muestras (genes) correladas y agrupamientos de muestras (genes) en los datos. 

###2.3. Expresin diferencial

Uno de los principales objetivos de muchos estudios de transcriptmica es encontrar genes que presentan un patrn de expresin diferencial entre distintos tipos de muestras. Por ejemplo, la identificacin de genes que muestran una expresin diferencial en pacientes respecto a controles sanos. 

En estos estudios se evalua si existe una diferencia significava en la media de expresin de cada gen en dos, o mas, condiciones analizadas. Dados un conjunto de muestras pertenecientes a dos grupos experimentales, A y B, como controles sanos y muestras tumorales, para cada gen se lleva a cabo un contrate de hiptesis

${H_0}: \mu_A=\mu_B$

${H_1}: \mu_A \neq\mu_B$

Aplicando el test correspondiente, se asigna a cada gen un *p-valor* que se usa para seleccionar aquellos que muestran una diferencia significativa en la expresin entre las dos condiciones. 

Existe mucha bibliografa respecto a los anlisis de expresin diferencial. Para introduciros en el tema, podis consultar la referencia Costa-Silva et al., 2017 de la bibliografa. 

El siguiente cdigo ilustra cmo aplicar un anlisis de expresin diferencial (DEA) entre muestras de tipo *Metstasis* y las de tipo *Tumor Slido Primario* utilizando la funcin `TCGAanalyze_DEA` sobre los datos descargados anteriormente del proyecto TCGA-SKCM. 

````
%%R

#Informacin de muestras
sample.info<-colData(SKCM.counts)
# Separar los datos en melanoma y tumor slido primario.
TMdata<-dataTMMnorm[,which(sample.info@listData[["definition"]]=="Metastatic")]
PSTdata<-dataTMMnorm[,which(sample.info@listData[["definition"]]=="Primary solid Tumor")]

# Anlisis de expresin diferencial entre metstasis y tumores slidos primarios.
dataDEGs <- TCGAanalyze_DEA(mat1 = TMdata,
                        	mat2 = PSTdata,
                        	Cond1type = "Metastatic",
                        	Cond2type = "Primary solid Tumor",
                        	fdr.cut = 0.01 ,
                        	logFC.cut = 1,
                        	method = "glmLRT") 

genesexpresadosdif <- as.character(rownames(dataDEGs))
genesexpresadosdif[1:10]
````

El resultado de estos anlisis es una lista de genes diferencialmente expresados, candidatos a ser estudiados en profundidad para identificar biomarcadores. ## **1. QU ES EL ANLISIS DE ENRIQUECIMIENTO DE ANOTACIONES FUNCIONALES?** <a id='funcional'></a>





El resultado del anlisis de datos omicos consiste, en la mayora de las ocasiones, en grandes listas de genes o protenas que estn asociados a un determinado fenotipo, por ejemplo, genes diferencialmente expresados entre dos condiciones (pacientes Vs sanos), genes con patrones de metilacin comunes, etc. A partir de este punto, el objetivo del estudio se centra en extraer conocimiento biolgico a partir de estos genes. Este conocimiento ayudar a entender cules son los procesos y funciones biolgicas desreguladas y servir como punto de partida para conocer las bases moleculares de los fenotipos que se estn estudiando.

Para ello, una de las aproximaciones clsicas ha consistido en evaluar si hay algn tipo de informacin (en forma de anotaciones) que est sobre-representada en la lista de genes respecto el resto de genes que hay en el genoma. Este tipo de anlisis, denominado anlisis funcional o anlisis de enriquecimiento, se basa en anotar los genes con informacin disponible en diferentes bases de datos como Gene Ontology (GO: http://geneontology.org/) o Kyoto Encyclopedia of Genes and Genomes (KEGG: https://www.genome.jp/kegg/), y establecer las frecuencias de cada trmino en la lista de genes y el resto del genoma, de forma que se puede aplicar un test estadstico para determinar qu anotaciones funcionales estn enriquecidas de forma significativa en la lista.

## **2. ANLISIS DE ENRIQUECIMIENTO DE ANOTACIONES INDIVIDUALES** <a id='single'></a>



El anlisis de enriquecimiento se basa en ver si hay una sobrerrepresentacin de una determinada anotacin en la lista de genes de inters respecto al resto del genoma. Uno de los test estadsticos usados con ms frecuencia en este contexto es el basado en la distribucin hipergeomtrica, donde para cada anotacin se puede calcular la probabilidad de encontrar un determinado nmero de genes asociados a la misma:

$P(X=i)={\frac{\binom{M}{i}\binom{N-M}{N-i}}{\binom{N}{n}}}$

Donde $N$ es el nmero total de genes en el genoma, $M$ es el nmero que presentan una determinada anotacin, $n$ es el nmero de genes en la lista e $i$ el nmero de genes en la lista con la anotacin.

Se calcula el p-valor, pero hay que tener en cuenta que cuando se analizan numerosos genes el p-valor se tiene que corregir para tener en cuenta el problema de las comparaciones mltiples, para lo que se suele aplicar correccin de Bonferroni o False Discovery Rate (FDR) entre las ms frecuentes.

## **4. GENE SET ENRICHMENT ANALYSIS (GSEA)**<a id='gsea'></a>



Gene Set Enrichment Analysis (Subramanian et al., 2005) es un algoritmo desarrollado para paliar algunas limitaciones de los anlisis de enriquecimiento que se aplican tras seleccionar una lista de genes diferencialmente expresados como: (i) en algunas ocasiones los anlisis de expresin diferencial no hay genes que superen los umbrales de significancia estadstica (ii) pueden existir genes que aunque no pasen los umbrales queden con valores muy cercanos y pueden aportar informacin muy til para la interpretacin funcional.

GSEA se basa en ordenar toda la lista de genes en base a la expresin diferencial entre dos condiciones sin aplicar un umbral para seleccionar un listado y evaluar la distribucin de los genes asociados a una determinada anotacin a lo largo de toda la lista y calcular un Enrichment Score. Los detalles sobre la metodologa se pueden encontrar en la publicacin original, pero de forma resumida:


- Dada una matriz *NxM* se ordenan los *N* genes en base a una medida de asociacin con el fenotipo $r(g_j)=rj$ obteniendo la lista ordenada *L: {g1, . . . , gN}* y un valor *p*.
- Dado un conjunto independiente de *$N_H$* genes *S*, por ejemplo genes que estn anotados con la misma funcin, se evalua la fraccin de genes en *S* (*T*) ponderada por su correlacin y la fraccin de genes que no estn en *S* (*F*) que hay hasta una posicin dada *i* en *L*.


$P_T (S,i)=\displaystyle\sum_{g_j\in S; j\leqslant i} ={\frac{ |{r_j}|^p }{N_R}}$, 

donde $N_R=\displaystyle\sum_{g_j\in S}|{r_j}|^p$

$P_F (S,i)=\displaystyle\sum_{g_j\notin S; j\leqslant i} ={\frac{ 1 }{(N-N_H)}}$

El *enrichment score* ES es la mxima desviacin de cero de $P_T-P_F$. 
Si S se distrbuye  aleatoriamente ES(S) tendr un valor pequeo, pero si se concentra en la parte superior o inferior de la lista ES(S) tendr un valor alto. 

Se calcula la significancia  calculando el ES obtenido al permutar las clases originales, repitiendo este proceso un gran nmero de veces y comparando el ES observado con los obtenidos en las permutaciones y corrigiendo finalmente los *p-valores* obtenidos por las comparaciones mltiples.
